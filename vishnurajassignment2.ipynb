```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/archive.zip')
df.head()
```

```python
sns.displot(df['Built Year'])
```

```python
sns.distplot(df['number of bedrooms'])
```

```python
plt.pie(df['number of floors'].value_counts(),[0.1,0,0.1,0,0.2,0.15],labels=['One','Two','One&half','Three','Two&half','Three&half'],autopct='%1.1f%%',shadow=True, colors=['red','blue','orange','green','yellow','pink'])
plt.title('Number Of Floors')
plt.show()
```

```python
sns.lineplot(x=df['number of floors'],y=df['Price'])
```

```python
sns.scatterplot(x=df['number of floors'],y=df['Price'])
```

```python
sns.heatmap(df.corr(),annot=True);
```
